```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
opioid_deaths = pd.read_csv('/Users/user/Documents/CS210/opioid_deaths.csv', engine='python')
admission_data = pd.read_csv('/Users/user/Documents/CS210/admissions.csv', engine='python')
nonmed_opioid_use = pd.read_csv('/Users/user/Documents/CS210/nonmed_pain_relievers.csv', engine='python')
```

```python
df = pd.DataFrame(opioid_deaths)

heatmap_data = opioid_deaths.pivot(index='state', columns='year', values='deaths')

plt.figure(figsize=(12, 10))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='.1f', linewidths=.5)
plt.title('Opioid Deaths by State (1999-2014)')
plt.xlabel('Year')
plt.ylabel('State')
plt.show()
```

It is evident that opioid use has multiplied over the years, with 2014 leading to the highest surge in opioid deaths from 1999 to 2014.

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='state', y='small_area_est', data=nonmed_opioid_use, errorbar=None)
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Small Area Estimate (%)')
plt.title('Small Area Estimates of Nonmedical Opioid Usage by State, 2012 to 2014')
plt.tight_layout()
plt.show()
```

This data makes it apparent that the region most impacted by opioid use in the U.S. is the West, which is in accordance with the previous heatmap. Oregon, in the West, had the highest number of opioid deaths in the same time frame.

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='number_2012', y='primary_substance', data=admission_data, color='b', label='2012')
sns.barplot(x='number_2013', y='primary_substance', data=admission_data, color='g', label='2013')
sns.barplot(x='number_2014', y='primary_substance', data=admission_data, color='r', label='2014')
plt.xlabel('Number of Instances')
plt.ylabel('Primary Substance')
plt.title('Total Counts of Primary Substances by Year')
plt.legend()
plt.tight_layout()
plt.show()
```

Interestingly, it can be seen that admissions into rehabilitation centers, hospitals, or other care facilities for opioid users generally decreased from 2012 to 2014.

This data underscores the increasing opioid problem in the US. While opioid-related deaths are steadily increasing, admission into care centers are decreasing. Certain regions are evidently more at-risk of opioid use, such as Western states like California and Oregon, and states in the South and Northeast such as Texas and New York, respectively.
